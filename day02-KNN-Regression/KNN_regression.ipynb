```python
df.drop(["sample_id","harvest_date"],axis = 1 , inplace = True)
```

```python
x = df.drop(["quality_score"],axis =1)
```

```python
from sklearn.compose import ColumnTransformer 
from sklearn.preprocessing import OrdinalEncoder , StandardScaler
ct = ColumnTransformer(transformers = [('oe',OrdinalEncoder(),[0,1,2,4]),
                                       ('sc',StandardScaler(),[3,5,6,7,8,9,10,11,12,])])
                      
```

```python
x_transformed  = ct.fit_transform(x)
x_transformed 
```

```python
x_transformed = pd.DataFrame(x_transformed , columns = x.columns)
```

```python
from sklearn.feature_selection import mutual_info_regression
mf = mutual_info_regression(x_transformed, y)
```

```python
mf = pd.Series(mf)
mf.index = x_transformed.columns
mf.sort_values(ascending = False)
```

```python
x.drop(["variety","region","tree_age_years"],axis = 1 , inplace = True)
```

```python
x.info()
```

```python
ct = ColumnTransformer(transformers = [('oe',OrdinalEncoder(),[0,2]),
                                      ('sc',StandardScaler(),[1,4,3,5,6,7,8,9])])
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2)
```

```python
x_train_transform = ct.fit_transform(x_train)
x_train_tansform = pd.DataFrame(x_train_transform)
```

```python
x_test_transform = ct.transform(x_test)
x_test_tansform = pd.DataFrame(x_test)
```

```python
from sklearn.neighbors import KNeighborsRegressor
kn = KNeighborsRegressor(n_neighbors = 7,p=1)
```

```python
kn.fit(x_train_transform,y_train)
```

```python
y_pred = kn.predict(x_test_transform)
```

```python
from sklearn.metrics import r2_score
r2 = r2_score(y_test, y_pred) * 100
print(r2)
```

```python
y_trian_pred = kn.predict(x_train_tansform)
```

```python
r = r2_score(y_train,y_trian_pred)
r
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsRegressor
kn = KNeighborsRegressor()

param_grid = {
    'n_neighbors': [3,5,7,9],
    'weights': ['uniform','distance'],
    'p': [1,2,3]
}

grid = GridSearchCV(estimator = kn , param_grid = param_grid , cv = 5 , scoring = 'r2' , verbose = 1)
grid.fit(x_train_transform,y_train)
```

```python
print("Best Parameters:",grid.best_params_)
print("Best CV Accuracy:",grid.best_score_)
```

```python
from sklearn.model_selection import RandomizedSearchCV
random_search = RandomizedSearchCV(
    kn,
    param_distributions = param_grid,
    n_iter = 30,
    cv= 5,
    scoring = 'r2',
    verbose = 1 , n_jobs = -1
)
random_search.fit(x_train_transform , y_train)
```

```python
print("Best Parameters:",random_search.best_params_)
print("Best CV Accuracy:",random_search.best_score_)
```

```python
from sklearn.model_selection import cross_val_score
score  = cross_val_score(kn,x_train_transform,y_train,cv = 5)
score
```

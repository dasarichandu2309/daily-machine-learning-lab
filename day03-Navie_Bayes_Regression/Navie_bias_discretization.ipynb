```python
import pandas as pd
import numpy as np
```

# Equal Width Binning

```python
# Implementation of Equal Width Binning
from sklearn.preprocessing import KBinsDiscretizer

x_continous = np.arange(1,101).reshape(10,10)
print(x_continous)
print("\n")
# using uniform strategy for equal_width bins
discretizer = KBinsDiscretizer(n_bins = 4 , encode = 'ordinal' , strategy = 'uniform')
x_binned = discretizer.fit_transform(x_continous)
x_binned
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.preprocessing import KBinsDiscretizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,classification_report
```

```python
data = load_iris()
x = pd.DataFrame(data.data,columns=data.feature_names)
y = data.target
```

```python
def discretize_data(x,strategy):
    kb = KBinsDiscretizer(n_bins=5,encode='ordinal',strategy=strategy)
    x_binned = kb.fit_transform(x)
    return pd.DataFrame(x_binned,columns = x.columns)

#compare Strategies
strategies = ['uniform','quantile','kmeans']
results = {}
for strat in strategies :
    x_binned = discretize_data(x,strat)
    x_train,x_test,y_train,y_test = train_test_split(x_binned,y,test_size= 0.3,random_state=42) 

    model = MultinomialNB()
    model.fit(x_train,y_train)
    y_pred = model.predict(x_test)

    acc = accuracy_score(y_test,y_pred)
    results[strat] = acc
    print(f"strategy: {strat}")
    print(classification_report(y_test,y_pred))

# plot results
plt.bar(results.keys() , results.values() ,color = 'skyblue')
plt.title("Naive Bayes Accuracy by Discretization Strategy")
plt.ylabel('Accuracy')
plt.show()
```
